```python
#import neccessary libraries
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
#load the dataset
df=pd.read_csv(r"C:\Users\kumar\Academic_Decision_Support_System\data\raw\student-mat.csv",sep=';')
df.head()
```

```python
# PASS RATE ANALYSIS
df["pass_fail"] = df["G3"].apply(lambda x: "Pass" if x >= 10 else "Fail")

df["pass_fail"].value_counts()
```

```python
# Select Important Features
features = ["studytime", "absences", "famsup", "schoolsup","G1","G2"]
X = df[features]
y = df["pass_fail"]

X.head()
```

```python
# encoding Categorical Variables

le = LabelEncoder()

X["famsup"] = le.fit_transform(X["famsup"])
X["schoolsup"] = le.fit_transform(X["schoolsup"])

X.head()
```

```python
# Save the cleaned dataset
cleaned_path = "C:\\Users\\kumar\\Academic_Decision_Support_System\\data\\processed\\student_performance_cleaned.csv"

final_df = X.copy()
final_df["pass_fail"] = y

final_df.to_csv(cleaned_path, index=False)

print("Saved cleaned dataset to:", cleaned_path)
```
