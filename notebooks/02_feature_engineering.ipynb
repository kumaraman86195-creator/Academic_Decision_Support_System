```python
#import neccessary libraries
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
#load the dataset
df=pd.read_csv(r"C:\Users\kumar\Academic_Decision_Support_System\data\raw\student_performance_balanced.csv")
df.head()
```

```python
# encoding Categorical Variables

label_cols = [
    "stress_level",
    "motivation_level",
    "family_support",
    "school_support",
    "activity_level"
]

encoders = {}

for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
#Encode Target Variable
df["pass_fail"] = df["pass_fail"].map({
    "Fail": 0,
    "Pass": 1
})
```

```python
# Select Features and Target
features = [
    "study_hours_per_day",
    "attendance_percentage",
    "absences",
    "sleep_hours",
    "stress_level",
    "motivation_level",
    "family_support",
    "school_support",
    "activity_level",
    "avg_grade"
]

X = df[features]
y = df["pass_fail"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Save the cleaned dataset
cleaned_path = "C:\\Users\\kumar\\Academic_Decision_Support_System\\data\\processed\\student_performance_cleaned.csv"


final_df = pd.concat([X, y], axis=1)
final_df.to_csv(cleaned_path, index=False)


print("ML-ready dataset saved to:", cleaned_path)
```
